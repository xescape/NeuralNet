```python
import pandas as pd, numpy as np
from pathlib import Path
import traceback
```

```python
working_dir = Path('/d/data/plasmo/nat_out/v4')
results_dir = Path('/d/data/plasmo/nn_scalar3')
ppi_df = pd.read_csv(working_dir / 'ppi.txt', sep='\t', index_col=1)
interesting_cols = ['Cluster#', 'Product Description', 'Symbol']
ppi_df = ppi_df.loc[:,interesting_cols]

hits_df = pd.read_csv(results_dir / 'output_mod' / 'predicted_genes_mod.tsv', sep='\t')
```

```python
#gets the cluster from the cluster table
def apply_fn(row):

    genes = [x.split(':')[0] for x in row['genes'].split(',')]
    clusters = []
    for g in genes:
        try:
            cluster = ppi_df.loc[g, 'Cluster#']
            try:
                cluster.shape[0]
                txt = '{0}: {1}'.format(g, ','.join([str(x) for x in cluster]))
            except:
                txt = '{0}: {1}'.format(g, cluster)
        except KeyError as e:
            print('{0} not found'.format(g))
            txt = '{0}: {1}'.format(g, 'NA')
            
        clusters.append(txt)
    return ';'.join(clusters)

hits_df['clusters'] = hits_df.apply(apply_fn, axis=1)
```

```python
hits_df.to_csv(results_dir / 'output_mod' / 'genes_with_clusters.tsv', sep='\t', index=False)
```
