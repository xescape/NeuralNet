```python
import pandas as pd
from pathlib import Path
```

```python
base_path = Path('/d/data/plasmo/')
```

```python
meta_path = base_path / 'SraRunTable.tsv'
meta_df = pd.read_csv(meta_path, sep='\t', header=0, usecols=['Run', 'Sample_Name'])
meta_df['name'] = meta_df['Sample_Name'].apply(lambda x: x[22:] if x.startswith('Plasmodium') else x)
meta_df = meta_df.set_index('Run').drop('Sample_Name', axis=1)
```

```python
meta_df
```

```python
used_path = base_path / 'training_data' / 'meta_v2.txt'
used_df = pd.read_csv(used_path, sep='\t', header=0)
used_df['id'] = used_df['name'].apply(lambda x: x[:-3])
used_df = used_df.set_index('id').drop('name', axis=1)
```

```python
res = used_df.merge(meta_df, how='inner', left_index=True, right_index=True)
```

```python
res.to_csv('/d/data/plasmo_training_1_names.csv')
```

```python
ts2_path = base_path / 'nat_out/v3/nat_v3.tsv'
with open(ts2_path, 'r') as in_file:
    names = in_file.readline().strip().split('\t')
```
