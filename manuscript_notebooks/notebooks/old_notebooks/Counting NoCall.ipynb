```python
import re, pandas as pd, numpy as np
from pathlib import Path

dir = Path('/d/data/plasmo/nat_out/v3')
in_path = dir / 'variants_v3.tsv'
out_path = dir / 'nocall.tsv'

```

```python
df = pd.read_csv(in_path, sep='\t', index_col = [0,1])
```

```python
def countRow(row):
    msk1 = row.values == '.'
    msk2 = row.values == '*'
    return sum(msk1) + sum(msk2)
counts = df.apply(countRow, axis = 1)
```

```python
counts.to_csv(out_path, sep = '\t')
```
