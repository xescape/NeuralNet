```python
import numpy as np, pandas as pd
import re

from pathlib import Path
```

```python
dir = Path("/d/data/toxo/old_toxo/Toxo/BaseData/tempz/similarity_results")
input = dir / "ME49.persistentMatrix.txt.sim"
with open(input, 'r') as i:
    data = i.read()
```

```python
split_data = [re.split("#", x) for x in re.split("@", data)[1:]] #nested list. first element is the chr name.
data_dict = {e[0].strip("\n"):[re.split("\n", i)[1:-1] for i in e[1:]] for e in split_data} #further formatting into nested list.
```

```python
data_dict.keys()
data_dict['TGME49_CHRIA']
```

```python
#this section finds percent white regions.
def f(x):
    if np.sum([float(re.split(":", e)[1].strip()) for e in x]) >= global_cutoff:
        return 1
    else:
        return 0

global_cutoff = 2.4
total = np.sum([f(x) for x in data_dict['TGME49_CHRIA']])
print(total, ' out of ', len(data_dict['TGME49_CHRIA']))
```

```python
#this section finds percent unique regions. 1 is less than global_cutoff for any of the other similarities
def f(x):
    vals = [float(re.split(":", e)[1].strip()) for e in x]
    del vals[global_self]
    if max(vals) <= global_cutoff:
        return 1
    else:
        return 0
    
global_self = 0
global_cutoff = 0.25

d = []
for e in data_dict.values():
    d.extend(e)
res = [f(x) for x in d]
total = np.sum(res)
print(total, ' out of ', len(res))
```

```python
#this section looks for specific similarity
#this section finds percent unique regions. 1 is less than global_cutoff for any of the other similarities
def f(x):
    vals = [float(re.split(":", e)[1].strip()) for e in x]
    if vals[global_target] >= global_cutoff:
        return 1
    else:
        return 0
    
global_target = 1
global_cutoff = 0.75

d = []
for e in data_dict.values():
    d.extend(e)
res = [f(x) for x in d]
total = np.sum(res)
print(total, ' out of ', len(res))
```
