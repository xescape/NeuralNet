```python
#CV of prediction accuracy
df = pd.read_csv((working_dir / 'snps_v3_base.fam'), sep=' ', header=None)
```

```python
def mae(x, y):
    return np.mean(np.abs(x - y), axis = None)

for x in range(10):
    pred_df = pd.read_csv((working_dir / 'preds_{0}.txt'.format(x)), sep=' ', header=None, index_col=None)
    df_pred = df.copy()
    df_pred['preds'] = pred_df[0]
    df_predonly = df_pred.dropna(axis=0, how='any')
    not_na = df_predonly[5] > -9
    df_predonly = df_predonly.loc[not_na, :]
    error = mae(df_predonly[5], df_predonly['preds'])
    expected_error = mae(df_predonly[5], np.full_like(df_predonly[5], np.mean(df[5], axis=None)))
    print(expected_error)
```
