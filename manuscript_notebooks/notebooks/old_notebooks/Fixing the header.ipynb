```python
import numpy as np, pandas as pd, re
from pathlib import Path

working_dir = Path('/d/data/plasmo/nn_scalar3')
in_file = working_dir / 'nat_nn.tsv'
out_file = working_dir / 'nat_nn_fixed.tsv'
```

```python
df = pd.read_csv(in_file, sep='\t', index_col=0)
```

```python
def zzz(col_names):
    new_names = []
    c = 0
    sl = 5000
    chr = ''
    for name in col_names:
        this_chr, this_pos = name.split(':')
        if this_chr == chr:
            new_names.append('{0}:{1}'.format(this_chr, int(sl * c)))
            c += 1
        else:
            new_names.append('{0}:{1}'.format(this_chr, 0))
            chr = this_chr
            c = 1
    return new_names

new_names = zzz(df.columns)
df.columns = new_names
df.to_csv(out_file, sep='\t')
        
```
