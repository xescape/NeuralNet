```python
'''
The difference between the popnet data and the sift data is a matter of position.
Once we have the snps sorted into genes, it's helpful to reset the index
to ranges of 1 to 5000 per gene in order to fit them into chunks for popnet
Then we should output a list of which chunk represent which gene.
'''
import numpy as np, pandas as pd, re
from pathlib import Path

in_path = Path('/d/data/plasmo/nat_out/v4') #path to input file

if not in_path.is_dir():
    in_path = Path('~/scratch/plasmo/op2')

in_file = in_path / 'combined_snps.tsv'
out_file = in_path / 'popnet_data.tsv'
genes_out_file = in_path / 'genes.tsv'
```

```python
df = pd.read_csv(in_file, sep = '\t', index_col = 0)
```

```python
SECTION_LENGTH = 5000
def changeIndex(args):
    idx = args[0]
    gene = args[1][0]
    section = args[1][1]
    n = section.shape[0]
    offset = idx * SECTION_LENGTH
    section['POS'] = range(offset + 1, offset + n + 1)
    return (section, gene)

groupby = df.groupby(level=0)
tmp = list(map(changeIndex, enumerate(groupby)))
section_dfs = [x[0] for x in tmp]
genes = [x[1] for x in tmp]
```

```python
res = pd.concat(section_dfs, axis=0)
res = res.reset_index()
res.loc[:, 'CHROM'] = 'Pf3D7_CHRI'
res.to_csv(out_file, sep='\t', index=False)
```

```python
with open(genes_out_file, 'w') as output:
    out_str = '\n'.join(genes)
    output.write(out_str)
```

```python
res.to_csv(out_file, sep='\t', index=False)
```
