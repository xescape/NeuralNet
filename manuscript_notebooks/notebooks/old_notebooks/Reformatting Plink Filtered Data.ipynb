```python
'''
This script was used to format data that
was filtered through PLINK in order to achieve some sort of
comparison with FastLMM. Even though we filtered the same way
AFAIK, I still ended up with about 2X the amount of SNPs even though
they had more starting samples. Nonetheless, I argue that it's
comparable enough... 
'''


import pandas as pd, numpy as np, re
from pathlib import Path

working_dir = Path('/d/data/plasmo/op3')
in_file = working_dir / 'variants_plink.tsv'
```

```python
df = pd.read_csv(in_file, sep='\t')
```

```python
def f_rename(x):
    p = '_(.+?)[.]GT'
    s = re.search(p, x)
    if s:
        return s.group(1)
    else:
        return x
    

df_renamed = df.rename(columns = f_rename)
```

```python
def f_apply(x):
    p = '[/](.+?)'
    try:
        s = re.search(p, x)
        if s:
            return s.group(1)
    except:
        return x

df_final = df_renamed.applymap(f_apply)
```

```python
def f_renameIndex(x):
    if x < 14:
        return 'Pf3D7_{0:0>2}_v3'.format(x + 1)
    elif x == 14:
        return 'Pf3D7_API_v3'
    elif x == 15:
        return 'Pf_M76611'
    else:
        raise Exception('hey a chromosome is too high {0}'.format(x))
    


df_zzz = df_final.set_index('CHROM')
df_zzz = df_zzz.rename(index = f_renameIndex)
```

```python
out_file = working_dir / 'nat_v3.tsv'
df_zzz.to_csv(out_file, sep='\t')
```
