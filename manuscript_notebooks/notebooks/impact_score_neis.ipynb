The neisseria version of the manhattan impact score graph should be easier since it doesn't need to be recalculated..

```python
from pathlib import Path
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
d_path = Path('/mnt/d/data/popnet_paper/')
data = pd.read_csv(d_path / 'neis_impact_score.tsv', sep='\t').sort_values(by='idx')
data['min_impact'] = data[['model_0', 'model_1', 'model_2', 'model_3', 'model_4']].abs().min(axis=1)
data['median_impact'] = data[['model_0', 'model_1', 'model_2', 'model_3', 'model_4']].abs().mean(axis=1)
data['position'] = data['id'].apply(lambda x: int(x.split(':')[1]))
# plot_data = data[['position', 'min_impact', 'median_impact']].melt(id_vars='position', value_vars=['total_impact_min', 'total_impact_avg'], var_name='impact_type', value_name='impact')
```

```python
# make a subplot with two rows, one for min and one for avg
sig_list = [747000, 745000, 1230000]
data['is_sig'] = data['position'].apply(lambda x: 'red' if x in sig_list else 'blue')
fig = make_subplots(rows=2, cols=1, subplot_titles=['median_impact', 'min_impact'])
fig.add_trace(go.Scatter(x=data['position'], y=data['min_impact'], mode='markers', name='min impact', marker=dict(size=4, color=data['is_sig'],)), row=2, col=1)
fig.add_trace(go.Scatter(x=data['position'], y=data['median_impact'], mode='markers', name='median impact', marker=dict(size=4, color=data['is_sig'],)), row=1, col=1)
fig.update_yaxes(title_text="impact_score", row=1, col=1,)
fig.update_yaxes(title_text="impact_score", row=2, col=1,)
fig.update_xaxes(title_text="position", row=1, col=1,)
fig.update_xaxes(title_text="position", row=2, col=1,)
fig.show()
```
